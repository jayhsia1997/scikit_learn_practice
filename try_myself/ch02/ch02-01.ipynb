```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Take 100 points between [0, 2 * pi]
x = np.linspace(0, 2 * np.pi, num=100)

# Calculate the sine of these 100 points and assign to the variable y
y = np.sin(x)
```

```python
# Draw x, y as a sine curve
plt.plot(x, y)
```
